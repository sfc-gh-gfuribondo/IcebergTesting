## Key Concepts

### Snowflake-Managed Iceberg Tables
- `CATALOG = 'SNOWFLAKE'` - Snowflake manages the Iceberg metadata
- `EXTERNAL_VOLUME` - Cloud storage location for data and metadata files
- `ICEBERG_VERSION = 3` - Use Iceberg v3 format for latest features

### Iceberg v3 Features
- Nanosecond timestamp precision (TIMESTAMP_NTZ(9))
- VARIANT column support for semi-structured data
- Default values for columns
- Multi-argument transforms
